#### Random stuff.

```python
y=Symbol('y')
f=(y+x)/(x-y)
display(f)
print(f.subs(x,-1).subs(y,2))
print(f.subs(x,3).subs(y,-1))
```

```python
f=65*x-180
display(f)
print(f.subs(x,1))
print(f.subs(x,2))
print(f.subs(x,3))
print(1680/65)
solve(f-1500,x)
print(336/13)
```

```python
f=x**2+y**2-4
display(f)
e1=Eq(f,3)
e2=Eq(f,4)
display(e1)
display(e2)
plot1=plot_implicit(e1,(x,-3.5,3.5),(y,-3.5,3.5))
plot2=plot_implicit(e2,(x,-3.5,3.5),(y,-3.5,3.5),line_color='red')
plot1.extend(plot2)
plot1.show()
```
